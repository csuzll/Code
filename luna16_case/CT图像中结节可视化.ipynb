### 显示结节

一个病人的一个CT检测出的肺结节有可能不止一个，可能分布于不同切片，有些结节位于同一张切片。
show_nodules()默认显示结节数为3。一般来讲，单个CT的肺结节数不会超过三个，所以只展示可能性最高的三个就应该够了。

plt.show()显示的图像：y代表纵轴，x代表横轴

````python
def show_nodules(ct_scan, nodules, origin, spacing, pad=2, max_show_num=3):
    """
    参数：
    ct_scan: 图像矩阵numpy(z,y,x)
    nodules:n * 4维的数组，n是肺结节数目，4代表x,y,z,以及直径。
    origin: 图像原点
    spacing: 图像像素间距
    pad：边的宽度
    max_show_num：最大展示数

    """
    show_index = []
#     numpyOrigin = np.array(list(ct_scan.GetOrigin())) # （x,y,z）
#     numpySpacing = np.array(list(ct_scan.GetSpacing()))  # (x,y,z)
    
#     numpyImage = sitk.GetArrayFromImage(ct_scan) # ct_scan的numpy数值矩阵
    
    for idx in range(nodules.shape[0]):
        if(idx < max_show_num):
            if(abs(nodules[idx,0])+abs(nodules[idx,1])+abs(nodules[idx,2])+abs(nodules[idx,3])==0): # 结节数为0就跳过下面的代码了。
                continue
            x = int(np.absolute((nodules[idx,0]-origin[0]))/spacing[0])
            y = int(np.absolute((nodules[idx,1]-origin[1]))/spacing[1])
            z = int(np.absolute((nodules[idx,2]-origin[2]))/spacing[2]) 

            print("结节体素坐标：",x,y,z) # 结节世界坐标转换为体素坐标

            radius = int(nodules[idx,3]/spacing[0]/2)

            data = ct_scan[z] # 结节所在的切片z

            #注意 y代表纵轴，x代表横轴
            data[max(0, y - radius):min(data.shape[0], y + radius), max(0, x - radius - pad):max(0, x - radius)] = 3000 # 竖线
            data[max(0, y - radius):min(data.shape[0], y + radius), min(data.shape[1], x + radius):min(data.shape[1], x + radius + pad)] = 3000 #竖线
            data[max(0, y - radius - pad):max(0, y - radius), max(0, x - radius):min(data.shape[1], x + radius)] = 3000 # 横线
            data[min(data.shape[0], y + radius):min(data.shape[0], y + radius + pad),max(0, x - radius):min(data.shape[1], x + radius)] = 3000 #横线

            if z in show_index: # 检查是否有结节在同一张切片，如果有，只显示一张
                continue
            show_index.append(z)

            plt.figure(idx)
            plt.imshow(data,cmap="gray")
    plt.show()
````

````python
# nodules为3个结节组成的array
nodules = np.array([[-116.2874457,21.16102581,-124.619925,10.88839157],[-111.1930507,-1.264504521,-138.6984478,17.39699158],[73.77454834,37.27831567,-118.3077904,8.648347161]])
print("结节世界坐标： ",nodules)

# 读入图像
itkimage = sitk.ReadImage(filename)#读取.mhd文件
numpyOrigin = np.array(list(itkimage.GetOrigin())) # （x,y,z）
numpySpacing = np.array(list(itkimage.GetSpacing()))  # (x,y,z)
Image = sitk.GetArrayFromImage(itkimage) # (z,y,x)

# 调用函数显示
show_nodules(Image, nodules, numpyOrigin, numpySpacing, pad=2,max_show_num=3)
````
