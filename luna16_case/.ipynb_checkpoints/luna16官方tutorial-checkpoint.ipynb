```python
import SimpleITK as sitk
import numpy as np
import csv
import os
from PIL import Image
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = "svg"
```

```python
# 定义一个函数读取图像，并转化为numpy，并提取图像相关信息
# return：numpyImage,numpyOrigin,numpySpacing
def load_itk_img(filename):
    itkimage = sitk.ReadImage(filename)
    numpyImage = sitk.GetArrayFromImage(itkimage) # 存的值为HU # (D,H,W)
    
    # 因为itkimage存储的与numpy图像存储的是反的
    numpyOrigin = np.array(list(reversed(itkimage.GetOrigin()))) # （z,y,x）
    numpySpacing = np.array(list(reversed(itkimage.GetSpacing())))  # (z,y,x)
    
    return numpyImage,numpyOrigin,numpySpacing
```

```python
# 定义一个函数读取csv文件
# return：csv文件的每一行
def readCSV(filename):
    lines = []
    with open(filename,"r") as f:
        csvreader = csv.reader(f)
        for line in csvreader:
            lines.append(line)
    return lines
```

```python
# 定义一个函数将the World Coordinates转换为voxel dates
# 非旋转的图像才可用以下方法

def worldToVoxelCoord(worldCoord,origin,spacing):
    stretchedVoxelCoord = np.absolute(worldCoord - origin)
    voxelCoord = stretchedVoxelCoord / spacing
    # 转换为整数voxel
    voxelCoord = voxelCoord.astype(np.int32)
    return voxelCoord
```
