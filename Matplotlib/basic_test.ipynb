```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 简单作图

x = np.linspace(-1,1,50) # 定义x：范围是(-1,1);个数是50
y = 2*x + 1

# plt.figure定义一个图像窗口
plt.figure()
plt.plot(x,y)  # 使用plt.plot画（x,y）曲线
plt.show()    #  plt.show显示图像
```

```python
# 同一窗口作多图

x = np.linspace(-3,3,50)
y1 = 2 * x +1
y2 = x ** 2

# 定义一个图像窗口：编号为3；大小为(8, 5)
plt.figure(num = 3,figsize=(8,5))
# 画(x ,y1)曲线，曲线的颜色(color)为红色;曲线的宽度(linewidth)为1.0；曲线的类型(linestyle)为虚线. 
plt.plot(x,y1,color="red",linewidth=1.0,linestyle="--")
# 画(x,y2)曲线
plt.plot(x,y2)
plt.show()
```

```python
# 设置坐标范围

plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# 设置坐标轴范围
plt.xlim((-1,2)) # 设置x坐标轴范围：(-1, 2)，默认0.5一个间隔
plt.ylim((-2,3)) # 设置x坐标轴范围：(-1, 2)
# 设置坐标轴名称
plt.xlabel("I am x") # 设置x坐标轴名称
plt.ylabel("I am y") # 设置y坐标轴名称
# 设置坐标轴刻度
my_x_ticks = np.linspace(-1, 2, 5) # 5个刻度
my_y_ticks = [-2,-1.8,-1,1.22,3] # 后面设置了对应刻度名称
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks,[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
# 挪动坐标轴位置
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y)
# 标注点（x0,y0)的位置信息
x0 = 1
y0 = 2*x0 + 1
# 第一个列表表示x的范围，第二个列表表示y的范围，k代表黑色，--代表虚线
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5) # 画出一条垂直于x轴的虚线
# 画散点（x0,y0）
plt.scatter([x0, ], [y0, ], s=50, color='b')
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
# 挪动坐标轴位置
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y)
# 标注点（x0,y0)的位置信息
x0 = 1
y0 = 2*x0 + 1
# 第一个列表表示x的范围，第二个列表表示y的范围，k代表黑色，--代表虚线
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5) # 画出一条垂直于x轴的虚线
# 画散点（x0,y0）
plt.scatter([x0, ], [y0, ], s=50, color='b')
# 添加注释 annotate
# 参数xycoords='data' 是说基于数据的值来选位置, xytext=(+30, -30) 和 textcoords='offset points' 对于标注位置的描述 和 xy 偏差值,
# arrowprops是对图中箭头类型的一些设置.
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
# 挪动坐标轴位置
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y)
# 标注点（x0,y0)的位置信息
x0 = 1
y0 = 2*x0 + 1
# 第一个列表表示x的范围，第二个列表表示y的范围，k代表黑色，--代表虚线
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5) # 画出一条垂直于x轴的虚线
# 画散点（x0,y0）
plt.scatter([x0, ], [y0, ], s=50, color='b')
# 添加注释 annotate
# 参数xycoords='data' 是说基于数据的值来选位置, xytext=(+30, -30) 和 textcoords='offset points' 对于标注位置的描述 和 xy 偏差值,
# arrowprops是对图中箭头类型的一些设置.
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
# 添加注释text
# 其中-3.7, 3,是选取text的位置, 空格需要用到转字符\ ,fontdict设置文本字体.
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})
plt.show()
```

```python
# 能见度
x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()
# 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
# zorder应该是图层的顺序
plt.plot(x, y, linewidth=10, zorder=1)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.show()
```

```python
# 设置 x轴 和 y轴 的刻度数字进行透明度设置
x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()
# 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
# zorder应该是图层的顺序
plt.plot(x, y, linewidth=10, zorder=1)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
# label.set_fontsize(12)重新调节字体大小
# bbox设置透明度相关
# facecolor调节 box 前景色，edgecolor 设置边框， 本处设置边框为无，alpha设置透明度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    # 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7, zorder=2))
plt.show()
```
