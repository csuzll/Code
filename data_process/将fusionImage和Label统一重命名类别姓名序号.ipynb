```python
import os
import shutil
```

```python
# 存放patient病人的CT目录,Fusion目录和Lable目录的地址

def make_dirpath(data_path):
    if(os.path.exists(data_path)):
        
        # 获取patient数据文件夹
        patient_filenames = os.listdir(data_path)
        
        # 存放label地址
        label_dirs = []
        # 存放patient的CT地址
        ct_dirs = []
        # 存放patient的PET地址
        fusion_dirs = []
        
        for patient_filename in patient_filenames:
            patient_filename_dir = os.path.join(data_path,patient_filename)
            t = os.listdir(patient_filename_dir)
            label_dirs.append(os.path.join(patient_filename_dir,t[2]))
            fusion_dirs.append(os.path.join(patient_filename_dir,t[1]))
            ct_dirs.append(os.path.join(patient_filename_dir,t[0]))
            
        return (label_dirs,ct_dirs,fusion_dirs)
    else:
        return "dataset path is not exit."
```

```python
datapath = "F:\\data\\yuanfazhao\\yuanfazhao_data_08"
lable_paths,patient_ct_paths,patient_fusion_paths = make_dirpath(datapath)
```

```python
print(patient_ct_paths)
print("\n")
print(patient_fusion_paths)
print("\n")
print(lable_paths)
```

```python
ct_all_path = []
fusion_all_path = []
label_all_path = []

for v in zip(patient_ct_paths,patient_fusion_paths,lable_paths):
    name = os.listdir(v[0])
    
    for n in name:
        ct_all_path.append(os.path.join(v[0],n))
        fusion_all_path.append(os.path.join(v[1],n))
        label_all_path.append(os.path.join(v[2],n))
```

```python
print(ct_all_path)
print("\n")
print(fusion_all_path)
print("\n")
print(label_all_path)
```

```python
# 重命名
for filename in zip(ct_all_path,fusion_all_path,label_all_path):
    w = filename[0].split("\\")
    w1 = w[2].split("-")
    
    new_name = w1[0]+w1[-1]+w[-1]
    
    os.rename(filename[0],os.path.join(os.path.split(filename[0])[0],new_name))
    os.rename(filename[1],os.path.join(os.path.split(filename[1])[0],new_name))
    os.rename(filename[2],os.path.join(os.path.split(filename[2])[0],new_name))
```
