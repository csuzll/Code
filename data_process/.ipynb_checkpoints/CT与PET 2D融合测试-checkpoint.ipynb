### 查看CT HU值的分布

pydicom读取到dicom存储的pixelData，最小值为-2000，是边界值，除此之外为最小值0。
-3024 = -2000 - 1024（-2000代表视野框以外，赋值-2000，不属于人体任何部分）

```python
# 读取ct并转换为HU,矩阵为ct_hu_slice[y,x]=>[H,W]
# 此时的ct_slice_arr为pixelData
ct_slice = pydicom.dcmread(ct_dir)
ct_slice_arr = ct_slice.pixel_array  # int16(-2000,0到4025)
print(ct_slice_arr.shape,ct_slice_arr.dtype,np.min(ct_slice_arr),np.max(ct_slice_arr))

ct_slice_arr = ct_slice_arr.astype(np.int16)
print(ct_slice_arr.shape,ct_slice_arr.dtype)
intercept = ct_slice.RescaleIntercept
slope = ct_slice.RescaleSlope
print("intercept:",intercept,"slope:",slope)
if slope != 1:
    ct_slice_arr = slope * ct_slice_arr.astype(np.float64)
    ct_slice_arr = ct_slice_arr.astype(np.int16)
# 此时的ct_slice_arr为HU值
    
ct_slice_arr += np.int16(intercept)

# ct_hu_slice为HU值，有负数
ct_hu_slice = ct_slice_arr # ct_hu_slice:int16(-1000多，到700左右)

plt.hist(ct_hu_slice.flatten(),bins=80,color="c")
plt.xlabel("Hounsfield Units(HU)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 显示HU值的图像
plt.imshow(ct_hu_slice,cmap=plt.cm.gray)
plt.show()
# 显示pixeldata的图像
ct_sli_arr = ct_slice.pixel_array
plt.imshow(ct_sli_arr,cmap=plt.cm.gray)
plt.show()
# 显示原图裁剪出的部分图像
ct_crop = ct_hu_slice[167:391,135:359]
plt.imshow(ct_crop,cmap=plt.cm.gray)
plt.show()
```

### 查看PET并重采样PET

将128的PET重采样与CT尺寸512一致

```python
# 读取PET
pet_slice = pydicom.dcmread(pet_dir)
pet_slice_arr = pet_slice.pixel_array # (H,W) int16(0左右，到1000多)
# 输出最大值和最小值
print(pet_slice_arr.shape,pet_slice_arr.dtype,np.min(pet_slice_arr),np.max(pet_slice_arr))
pet_slice_arr = pet_slice_arr.astype(np.int16)
print(pet_slice_arr.shape,pet_slice_arr.dtype)
print(pet_slice_arr[75][76])
plt.imshow(pet_slice_arr,cmap=plt.cm.gray)
plt.show()
```

## CT与PET小波融合

ct_hu_slice:  515 * 512  (HU值CT）

pet_img: 512 * 512 (转换后的重采样的PET）

```python
import pywt
```

```python
# 进行ct_hu_slice和pet_img的融合
ct = np.copy(ct_hu_slice).astype(np.float64) # ct
pet = np.copy(pet_img).astype(np.float64)

imagearraylist = [ct,pet]
print(imagearraylist[0].shape,imagearraylist[0].dtype)
print(imagearraylist[1].shape,imagearraylist[1].dtype)
```

```python
coeffss = [] # 每张图的小波分解系数列表组成的列表
zt = 2  # 设置小波分解的层数
mp = 0  # 低频融合规则：0-average, 1-min, 2-max
ap = 2  # 高频融合规则：0-average, 1-min, 2-max，3基于区域特性融合（区域匹配度）


# 构造此列表
for image in imagearraylist:
    coeffss.append(pywt.wavedec2(image, 'db1', level=zt))
    
# 低频，基于平均值融合融合
if mp == 0:
    cAF = coeffss[0][0]  # first image的近似系数 
    for coeffs in coeffss[1:]:   # coeffss[1:]就是second image分解系数列表组成的列表。
        cAF += coeffs[0]  # cAF更新为加上second image的近似系数
        cAF = cAF/2  # cAF更新为取平均值后的融合近似系数,??????,为什么时除以len(coeffs)，应该是除以2啊。

# 高频，系数绝对值较大法
if ap == 2:
    hipassF  = coeffss[0][1:] # first image的各种细节系数
    for coeffs in coeffss[1:]:   # every image,这里特指second image
        for idxLevel, HVDs in enumerate(coeffs[1:]):   # every level
            for idxDirec, HVD in enumerate(HVDs):     # 融合高频的各个系数
                maxMap = hipassF[idxLevel][idxDirec] < HVD
                hipassF[idxLevel][idxDirec][maxMap] = HVD[maxMap]
# 最后的融合系数
coeffsFusion = [cAF,] + hipassF

# 逆分解得到融合图像像素值矩阵
fusionImage = pywt.waverec2(coeffsFusion,"db1")
# 输出一些值看看
print(type(fusionImage),fusionImage.dtype,fusionImage.shape)
```

```python
print(np.min(fusionImage),np.max(fusionImage))
plt.imshow(fusionImage,cmap=plt.cm.gray)
plt.show()
```

```python
fusion_crop = fusionImage[147:371,135:359]
plt.imshow(fusion_crop,cmap=plt.cm.gray)
plt.show()
```

###  CT和PET加权融合

利用cv2.addWeighted（）函数来实现加权融合

```python
# 加权融合
dst=cv2.addWeighted(ct,0.7,pet,0.3,0.0)
print(dst.dtype)
plt.imshow(dst,cmap=plt.cm.gray)
plt.show()

# 裁剪224 * 224
dst_crop = dst[147:371,135:359]
plt.imshow(dst_crop,cmap=plt.cm.gray)
plt.show()

print(np.min(dst_crop),np.max(dst_crop))

plt.hist(dst_crop.flatten(),bins=80,color="c")
plt.xlabel("Hounsfield Units(HU)")
plt.ylabel("Frequency")
plt.show()

dst_crop_int = dst_crop.astype(np.int32)
print(np.min(dst_crop_int),np.max(dst_crop_int))
plt.imshow(dst_crop_int,cmap=plt.cm.gray)
plt.show()
```

### 将融合后值按缩放到[-1200,600]，并转换为0-255

```python
lungwin = np.array([-1200.,600.])
newimg = (dst_crop-lungwin[0])/(lungwin[1]-lungwin[0])
newimg[newimg<0]=0
newimg[newimg>1]=1
# 缩放到0~255
newimg = (newimg*255).astype('uint8')
plt.imshow(newimg,cmap=plt.cm.gray)
plt.show()

print(newimg.shape)

# 将单通道 224 * 224 的patch变为3通道，简单的复制3便即可
new_img_3c = np.stack((newimg,newimg,newimg))
print(new_img_3c.shape)  # C,H,W

np.save("fusion.npy",new_img_3c) # 保存为npy

c = np.load("fusion.npy")
print(c.shape)
print(c.dtype)
```
